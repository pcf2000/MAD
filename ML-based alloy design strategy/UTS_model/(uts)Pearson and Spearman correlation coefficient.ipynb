```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Wedge

# Excel
data = pd.read_excel('all_UTS.xlsx')

# 
corr = data.corr()

# 
fig, ax = plt.subplots(figsize=(12, 10))
ax.set_xlim(0, corr.shape[1])
ax.set_ylim(0, corr.shape[0])

# y
ax.set_yticks(np.arange(0.5, len(corr.columns), 1))
ax.set_yticklabels(corr.columns[::-1])

# x
ax.set_xticks([])

# 
for i in range(corr.shape[0]):
    for j in range(i + 1, corr.shape[1]):
        value = corr.iloc[i, j]
        size = abs(value)  # 
        if value > 0:
            color = plt.cm.Blues(size)  # Blues
        else:
            color = plt.cm.Reds(size)  # Reds
        # 
        bg_wedge = Wedge((j + 0.5, corr.shape[0] - i - 0.5), 0.4, 0, 360, facecolor='white', edgecolor='black', alpha=1)
        ax.add_patch(bg_wedge)
        # 
        wedge = Wedge((j + 0.5, corr.shape[0] - i - 0.5), 0.4, 90, 90 + 360 * size, facecolor=color, edgecolor='black', alpha=1)
        ax.add_patch(wedge)
        # 
        ax.text(j + 0.5, corr.shape[0] - i - 0.5, f'{value:.2f}', horizontalalignment='center', verticalalignment='center', fontsize=8)

# 
norm = plt.Normalize(-1, 1)
cbar = plt.colorbar(plt.cm.ScalarMappable(norm=norm, cmap='coolwarm'), ax=ax, orientation='vertical')
cbar.set_label('Pearson correlation coefficient')

plt.title('Pearson Correlation Coefficient Plot (Upper Triangle)')

# 
plt.show()
```
